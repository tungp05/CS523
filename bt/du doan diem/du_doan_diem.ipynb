```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
```

```python
file_path = "/content/drive/MyDrive/bt/TRAIN2.xlsx"
df = pd.read_excel(file_path)

print("Dataset:")
print(df.head())
```

```python
X = df[["midterm"]]
y = df["final"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = MLPRegressor(
    hidden_layer_sizes=(10, 10),
    activation='relu',
    solver='adam',
    max_iter=1000,
    random_state=42
)


model.fit(X_scaled, y)
```

```python
midterm_input = float(input("\\nNhap diem giua ky: "))

input_scaled = scaler.transform([[midterm_input]])

predicted = model.predict(input_scaled)

print(f"Diem cuoi ky du doan: {predicted[0]:.2f}")
```

```python
plt.plot(model.loss_curve_)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Neural Network Training Loss")

plt.show()
```
